```python
import pandas as pd
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding='latin')
df
```

```python
df = df[['PRICEEACH','SALES', 'QTR_ID', 'MONTH_ID', 'YEAR_ID','PRODUCTLINE','MSRP', 'COUNTRY', 'DEALSIZE', 'QUANTITYORDERED']]
```

```python
for col in df:
  if df[col].dtypes == 'object':
    print(f'\n{col} --> {df[col].unique()}')
```

```python
df = pd.get_dummies(data = df, columns = ['PRODUCTLINE', 'COUNTRY', 'DEALSIZE'])
df
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
sse = []
K = range(1,10)
for k in K:
  KMeansModel = KMeans(n_clusters = k)
  KMeansModel.fit(df)
  sse.append(KMeansModel.inertia_) 
```

```python
plt.figure(figsize = (16,8))
plt.plot(K, sse, 'o-')
plt.xlabel('k')
plt.ylabel('sse')
plt.title('Elbow Method')
plt.show()
```

```python
import numpy as np
```

```python
x_train = df.values
x_train
```

```python
KMeansModel = KMeans(n_clusters = 3)
KMeansModel = KMeansModel.fit(x_train)
predictions = KMeansModel.predict(x_train)
```

```python
unique, counts = np.unique(predictions, return_counts = True)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
reducedX = pd.DataFrame(pca.fit_transform(x_train), columns = ['PCA1', 'PCA2'])
reducedX.head()
```

```python
plt.figure(figsize = (14,10))
plt.scatter(reducedX['PCA1'], reducedX['PCA2'])
```

```python
# KMeansModel.cluster_centers_
reduced_centers = pca.transform(KMeansModel.cluster_centers_)
reduced_centers
```

```python
plt.figure(figsize = (14,10))
plt.scatter(reducedX['PCA1'], reducedX['PCA2'])
plt.scatter(reduced_centers[:, 0], reduced_centers[:,1], color='black',marker='x',s=300)
```
